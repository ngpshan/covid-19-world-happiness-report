```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, ScalarFormatter 

df = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
df.head()
```

```python
plt.plot(df['Year'], df['Math and Statistics'], c='red', label='Women') # c = color
plt.plot(df['Year'], 100-df['Math and Statistics'], c='blue', label='Men')

plt.legend(loc='upper right') # location of legend
plt.title('Percentage of Math Degrees Awarded By Gender')
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'], df['Computer Science'], c='red', label='Women')
ax.plot(df['Year'], 100-df['Computer Science'], c='blue', label='Men')
ax.tick_params(top=False, bottom=False, left=False, right=False, labelleft=True, labelbottom=True)


for key,spine in ax.spines.items(): # ax.spines["right"].set_visible(False) removes spines for right axes
    spine.set_visible(False)

ax.set_title('Percentage of Computer Science Degrees Awarded By Gender')
ax.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.show()
```

```python
degrees = ['Business', 'Computer Science', 'Foreign Languages', 'Math and Statistics']
```

```python
fig = plt.figure(figsize=(10, 10)) # fix size of plots

for s in range(0,4):
    ax = fig.add_subplot(2,2, s+1)
    ax.plot(df['Year'], df[degrees[s]], c='red', label='Women')
    ax.plot(df['Year'], 100-df[degrees[s]], c='blue', label='Men')

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
```

## (Graph below) No spines, with titles

```python
fig = plt.figure(figsize=(10, 10)) # fix size of plots

for s in range(0,4):
    ax = fig.add_subplot(2,2, s+1)
    ax.plot(df['Year'], df[degrees[s]], c='red', label='Women')
    ax.plot(df['Year'], 100-df[degrees[s]], c='blue', label='Men')

    for key,spine in ax.spines.items(): # removed spines in the graph below **
        spine.set_visible(False)

    ax.set_xlim(1968, 2011) # x axis limiting range of years 
    ax.set_ylim(0,100)      # y axis limiting from 0-100

    ax.set_title(degrees[s]) # added titles for plots **

    ax.tick_params(top=False, bottom=False, left=False, right=False, labelleft=True, labelbottom=True)

# Calling pyplot.legend() here will add the legend to the last subplot that was created.
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
```
